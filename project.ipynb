```python
import numpy as np
import pandas as pd

# Column names for years 2017-2021 we need to explore
columns = {
    'y17': ['YearsCodedJob', 'HaveWorkedLanguage', 'WantWorkLanguage'],
    'y18': ['YearsCodingProf', 'LanguageWorkedWith', 'LanguageDesireNextYear'],
    'y19': ['YearsCodePro', 'LanguageWorkedWith', 'LanguageDesireNextYear'],
    'y20': ['YearsCodePro', 'LanguageWorkedWith', 'LanguageDesireNextYear'],
    'y21': ['YearsCodePro', 'LanguageHaveWorkedWith', 'LanguageWantToWorkWith']
}
new_proportions = {}
```

```python
## For different years the year of experience values are different. Some year servey lets you chose number 4,5,6 ... 
## Another year you may have option to choose only range '4 to 5 years', OR '3-5 years'
## Hence it is difficult to compare year by year. But, by chance it is possible to group into <= 5 or more
## The following values are all that specify <= 5 years of experience
lessthan5 = ['1 to 2 years', '2 to 3 years', '3 to 4 years', '4 to 5 years', 'Less than a year', 'Less than 1 year',
             '0-2 years', '3-5 years', '1','2','3','4', '5']
lessthan5_regex = "^{}$".format('$|^'.join(lessthan5))
print("Regex used: ", lessthan5_regex)
```

## Proportions of "new" programmers
Proportion of programers with less than 5 years of experience and more

```python
df = pd.read_csv('survey/stack-overflow-developer-survey-2017/survey_results_public.csv')
df = df[columns['y17']]
df["LessThanFive"] = df['YearsCodedJob'].str.contains(lessthan5_regex)
new_proportions["2017"]  = df['LessThanFive'].mean()
```

## Trend

```python
trend =  pd.Series(new_proportions)
trend.head()
```

```python
import matplotlib.pyplot  as plt
fig, ax = plt.subplots(figsize=(12, 6))
ax.set(title='Proportion of programmers with less than 5 years experience',
      xlabel='Year', ylabel='Proportion')
ax.plot(trend, color='blue', label='Portion')
plt.ylim([0,1])
plt.grid()
plt.show()
fig.savefig('newDev.png',
            dpi=300, 
            format='png', 
            bbox_inches='tight')
```

## Popular language

```python
# Explore languages used in 2021
df = pd.read_csv('survey/stack-overflow-developer-survey-2021/survey_results_public.csv')
df = df[columns['y21']]
lang = df['LanguageHaveWorkedWith'].str.split(';', expand=True).stack()
lang.head(20)
```

```python
def getLanguageStat(filename, year, col):
    data = pd.read_csv(filename)
    df = data[columns[year]]
    df = df.dropna(subset=[col])
    length = df.shape[0]
    df[col] = df[col].str.replace('\s*;\s*', ';', regex=True)
    frequency = df[col].str.split(';', expand=True).stack().value_counts()
    
    return frequency/length
```

```python
fig, ax = plt.subplots(figsize=(12, 10))
filename = 'survey/stack-overflow-developer-survey-2021/survey_results_public.csv'
getLanguageStat(filename, 'y21', 'LanguageHaveWorkedWith')[0:15].plot(kind='barh', width=0.9)
ax.set(title="Top 15 Languages from 2021 survey", 
       xlabel="Proportion of responders that worked with a Language",
      xlim=[0, .75])
fig.savefig('languages2021.png',
            dpi=300, 
            format='png', 
            bbox_inches='tight')
```

```python
filename = 'survey/stack-overflow-developer-survey-2017/survey_results_public.csv'
s17 = getLanguageStat(filename, 'y17', 'HaveWorkedLanguage')[0:15]
filename = 'survey/stack-overflow-developer-survey-2018/survey_results_public.csv'
s18 = getLanguageStat(filename, 'y18', 'LanguageWorkedWith')[0:15]
filename = 'survey/stack-overflow-developer-survey-2019/survey_results_public.csv'
s19 = getLanguageStat(filename, 'y19', 'LanguageWorkedWith')[0:15]
filename = 'survey/stack-overflow-developer-survey-2020/survey_results_public.csv'
s20 = getLanguageStat(filename, 'y20', 'LanguageWorkedWith')[0:15]
filename = 'survey/stack-overflow-developer-survey-2021/survey_results_public.csv'
s21 = getLanguageStat(filename, 'y21', 'LanguageHaveWorkedWith')[0:15]
```

```python
df = pd.concat([s17,s18, s19, s20, s21], axis=1).dropna() 
df.head(20)
```

```python
df.rename(columns={0: 2017,1: 2018,2: 2019,3: 2020,4: 2021}, inplace=True)
df.head(20)
```

```python
fig, ax = plt.subplots(figsize=(12, 6))
ax.plot(df.transpose(), );
ax.set(title="Trend of the top Languages from 2017-2021", 
       xlabel="Year",
       ylabel='Proportion',
      xlim=[2017, 2021],
      ylim=[0, .8]
      )

plt.xticks((2017, 2018, 2019, 2020, 2021));
ax.legend(list(df.index), bbox_to_anchor=(0.995, 1.01), loc='upper left')
fig.savefig('languages_trend.png',
            dpi=300, 
            format='png', 
            bbox_inches='tight'
            )
```
